# UMaT Management System

Inheritance case study implementation.

```

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")


class Student(Person):
    def __init__(self, name, age, student_id, enroll_course):
        super().__init__(name, age)
        self.student_id = student_id
        self.enroll_course = enroll_course

    def add_student(self):
        print(f"Student {self.student_id} enrolled in {self.enroll_course}")

    def display_info(self):
        super().display_info()
        print(f"Student ID: {self.student_id}")
        print(f"Course: {self.enroll_course}")


class Lecturer(Person):
    def __init__(self, name, age, employee_id, teach_course):
        super().__init__(name, age)
        self.employee_id = employee_id
        self.teach_course = teach_course

    def add_employee(self):
        print(f"Lecturer {self.employee_id} teaches {self.teach_course}")

    def display_info(self):
        super().display_info()
        print(f"Employee ID: {self.employee_id}")
        print(f"Course Taught: {self.teach_course}")


student = Student("Ama Mensah", 20, "STU001", "Programming")
lecturer = Lecturer("Kwame Asare", 40, "EMP101", "Data Structures")

student.display_info()
print()
lecturer.display_info()
```
